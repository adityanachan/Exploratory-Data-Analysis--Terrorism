```python
null_values=data1.isnull().sum()
null_values_percentage=100*null_values/len(data1)
null_val=pd.concat([null_values,null_values_percentage],axis=1)
null_value=null_val.rename(columns={0:'null_values',1:'null_values_percentage'},inplace=True)
```

```python
null_val.head()
```

```python
data1['Motive'].fillna(value='NA', inplace=True) 
data1['Summary'].fillna(value='NA', inplace=True)
data1['city'].fillna(value='NA', inplace=True)
data1['Target'].fillna(value='NA', inplace=True)
data1["Killed"].fillna(data1["Killed"].mean(), inplace=True)
data1["Wounded"].fillna(data1["Wounded"].mean(), inplace=True)
data1["latitude"].fillna(data1["latitude"].mean(), inplace=True)
data1["longitude"].fillna(data1["longitude"].mean(), inplace=True)
data1['state'].fillna(value='NA', inplace=True)
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(data1.corr(),cmap="YlGnBu", annot=True, xticklabels='auto', yticklabels='auto', linewidth=1 )
```

```python
plt.subplots(figsize=(15,6))
sns.countplot('Year',data=data1,palette="viridis")
plt.xticks(rotation=45)
plt.title('Number Of Terrorist Activities Each Year')
plt.show()
```

```python
pd.crosstab(data1.Year, data1.Region).plot(kind='area',figsize=(15,6))
plt.ylabel('Number of Attacks')
plt.show()
```

```python
attack= data.Country.value_counts()[:10]
attack
```

```python
plt.subplots(figsize=(15,6))
sns.barplot(data['Country'].value_counts()[:15].index,data['Country'].value_counts()[:15].values,palette='Blues_d')
plt.title('Top Countries Affected')
plt.xlabel('Countries')
plt.ylabel('Count')  
plt.xticks(rotation=90)
plt.show()                                         
```

```python
data['city'].value_counts().to_frame().sort_values('city',axis=0,ascending=False).drop("Unknown").head(10).plot(kind='bar',figsize=(20,10),color='black')

plt.ylabel("Number of attack",fontsize=20)
plt.xlabel("City",fontsize=20)
plt.title("Top 10 most effected city", fontsize=28)
plt.show()
```

```python
data['AttackType'].value_counts().plot(kind='bar',figsize=(20,10),color='lightblue')
plt.ylabel("Number of attack",fontsize=20)
plt.xlabel("AttackType",fontsize=20)
plt.title("Name of AttackType", fontsize=28)
plt.show()
```

```python
data[['AttackType','Killed']].groupby(["AttackType"],axis=0).sum().plot(kind='bar',figsize=(20,10),color=['darkblue'])
plt.ylabel("Number of people",fontsize=20)
plt.xlabel("AttackType",fontsize=20)
plt.title("Killed Peoples", fontsize=28)
plt.show()
```

```python
data[['AttackType','Wounded']].groupby(["AttackType"],axis=0).sum().plot(kind='bar',figsize=(20,10),color=['darkgreen'])
plt.ylabel("Number of people",fontsize=20)
plt.xlabel("AttackType",fontsize=20)
plt.title("Wounded Peoples", fontsize=28)
plt.show()
```

```python
data['Group'].value_counts().to_frame().drop("Unknown").head(10).plot(kind='bar',color="y",figsize=(20,10))
plt.xlabel("Terrorist Group Name",fontsize=20)
plt.ylabel("Attack Number",fontsize=20)
plt.title("Top 10 terorist group ", fontsize=28)
plt.show()
```

```python
data[['Group','Killed']].groupby(["Group"],axis=0).sum().drop('Unknown').sort_values('Killed',ascending=False).head(10).plot(kind='bar',figsize=(20,10),color=['blue'])
plt.ylabel("No of killed people",fontsize=20)
plt.xlabel("Terrorist Group Name",fontsize=20)
plt.title("Top 10 terrorist group attack", fontsize=28)
plt.show()
```
